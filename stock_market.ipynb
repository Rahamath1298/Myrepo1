```python
!pip install finnhub-python

import finnhub
import time
from datetime import datetime

class StockTracker:
    def __init__(self, api_key):
        """Initialize the StockTracker with your API key."""
        self.client = finnhub.Client(api_key=api_key)
        self.valid_regions = {'US', 'GB', 'HK', 'IN'}

    def validate_api_key(self):
        """Test the API key by making a simple API call."""
        try:
            # Try to fetch a well-known stock as a test
            self.client.quote('AAPL')
            print("API connection successful!")
            return True
        except finnhub.FinnhubAPIException as e:
            if e.status_code == 401:
                print("\nERROR: Invalid API key!")
                print("Please check that you're using the correct API key.")
                print("Current API key:", self.client.api_key)
                print("\nTo get a valid API key:")
                print("1. Go to https://finnhub.io/")
                print("2. Create a free account")
                print("3. Get your API key from the dashboard")
            else:
                print(f"API Error: {e}")
            return False

    def is_valid_region(self, region):
        """Check if the region code is valid."""
        return region in self.valid_regions

    def fetch_symbols(self, region):
        """Fetch stock symbols for a given region with error handling."""
        if not self.is_valid_region(region):
            print(f"\nError: '{region}' is not a valid region code.")
            print("Please use one of these region codes:", ", ".join(sorted(self.valid_regions)))
            return []

        try:
            print(f"\nFetching symbols for {region}...")
            symbols = self.client.stock_symbols(region)
            if not symbols:
                print(f"No symbols found for region {region}")
                return []

            print(f"\nTotal symbols available in {region}: {len(symbols)}")
            print("\nSample symbols:")
            for symbol in symbols[:5]:
                print(f"Symbol: {symbol['symbol']:<15} Description: {symbol['description']}")
            return symbols

        except finnhub.FinnhubAPIException as e:
            if e.status_code == 401:
                print("Error: Authentication failed. Please check your API key.")
            else:
                print(f"Error fetching symbols: {e}")
            return []

    def get_stock_data(self, symbol):
        """Fetch and display real-time stock data with error handling."""
        try:
            stock_data = self.client.quote(symbol)

            if not stock_data or stock_data.get('c') == 0:
                print(f"\nNo valid data available for {symbol}.")
                print("Please check that you entered a valid stock symbol.")
                print("Example symbols: AAPL, MSFT, GOOGL, AMZN")
                return False

            # Format timestamp
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            # Calculate price change
            price_change = stock_data['d']
            price_change_percent = stock_data['dp']

            # Create output with colorful indicators (/)
            change_indicator = "" if price_change >= 0 else ""

            print("\n" + "="* 50)
            print(f"Time: {timestamp}")
            print(f"Stock: {symbol}")
            print(f"Current Price: ${stock_data['c']:.2f}")
            print(f"Daily Change: {change_indicator} ${abs(price_change):.2f} ({price_change_percent:.2f}%)")
            print(f"Day Range: ${stock_data['l']:.2f} - ${stock_data['h']:.2f}")
            print(f"Opening Price: ${stock_data['o']:.2f}")
            print(f"Previous Close: ${stock_data['pc']:.2f}")
            print("=" * 50)
            return True

        except finnhub.FinnhubAPIException as e:
            print(f"\nError fetching data for {symbol}: {e}")
            return False

def get_region_input():
    """Get and validate region input."""
    while True:
        print("\nAvailable Regions:")
        print("US - United States")
        print("GB - United Kingdom")
        print("HK - Hong Kong")
        print("IN - India")
        region = input("\nEnter region code (US, GB, HK, or IN): ").upper().strip()
        if region in {'US', 'GB', 'HK', 'IN'}:
            return region
        print(f"\nError: '{region}' is not a valid region code.")
        retry = input("Would you like to try again? (y/n): ").lower().strip()
        if retry != 'y':
            return None

def display_menu():
    """Display the main menu with clear instructions."""
    print("\n" + "="*50)
    print("Stock Market Tracker - Main Menu")
    print("="*50)
    print("Please select an option by entering a number:")
    print("1. Search symbols by region")
    print("2. Track specific stock")
    print("3. Exit")
    print("="*50)

def get_menu_choice():
    """Get and validate menu choice."""
    while True:
        try:
            choice = input("\nEnter your choice (1, 2, or 3): ").strip()
            if choice in {'1', '2', '3'}:
                return choice
            print("Error: Please enter 1, 2, or 3 only.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")

def main():
    # Your API key - make sure this is correct
    API_KEY = "cu2eolhr01ql7sc7bmbgcu2eolhr01ql7sc7bmc0"

    # Initialize tracker
    tracker = StockTracker(API_KEY)

    # Validate API key first
    if not tracker.validate_api_key():
        return

    while True:
        display_menu()
        choice = get_menu_choice()

        if choice == '1':
            region = get_region_input()
            if region:
                tracker.fetch_symbols(region)

        elif choice == '2':
            print("\nEnter a stock symbol to track.")
            print("Examples: AAPL (Apple), MSFT (Microsoft), TSLA (Tesla)")
            symbol = input("Enter stock symbol: ").upper().strip()
            print("\nPress Ctrl+C to stop tracking")
            try:
                while True:
                    if not tracker.get_stock_data(symbol):
                        input("\nPress Enter to return to main menu...")
                        break
                    time.sleep(10)  # Update every 10 seconds
            except KeyboardInterrupt:
                print("\nStock tracking stopped")

        elif choice == '3':
            print("\nExiting program...")
            break

if __name__ == "__main__":
    main()
```

```python
import requests
from time import sleep

def get_stock_data(symbol, api_key):
    """Fetch stock data for a given symbol using Twelve Data API"""
    base_url = 'https://api.twelvedata.com/time_series'
    params = {
        'symbol': symbol,
        'interval': '1min',  # You can adjust the interval as needed
        'apikey': api_key,
        'outputsize': 1  # Fetch the latest data point
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        data = response.json()
        if 'values' in data:
            return data['values'][0]  # Return the latest data point
        else:
            print(f"Error fetching data: {data.get('message', 'Unknown error')}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

def get_symbol_search(keywords, api_key):
    """Search for stock symbols based on keywords using Twelve Data API"""
    base_url = 'https://api.twelvedata.com/symbol_search'
    params = {
        'symbol': keywords,
        'apikey': api_key
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        data = response.json()
        if 'data' in data:
            return data['data']
        else:
            print(f"Error searching symbols: {data.get('message', 'Unknown error')}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error searching symbols: {e}")
        return None

def get_valid_input(prompt, valid_options):
    """Get and validate user input"""
    while True:
        user_input = input(prompt).strip()
        if user_input in valid_options:
            return user_input
        print(f"Invalid input. Please enter one of: {', '.join(valid_options)}")

def handle_symbol_search(api_key):
    """Handle the symbol search process"""
    while True:
        keywords = input("Enter company name to search (or 'back' to return): ").strip()
        if keywords.lower() == 'back':
            return None

        search_results = get_symbol_search(keywords, api_key)

        if search_results:
            print("\nSearch Results:")
            for idx, match in enumerate(search_results, 1):
                print(f"{idx}. Symbol: {match['symbol']} - {match['instrument_name']} ({match['exchange']})")

            while True:
                symbol_choice = input("\nSelect a symbol number (or 'back' to search again): ").strip()
                if symbol_choice.lower() == 'back':
                    break

                try:
                    choice_idx = int(symbol_choice) - 1
                    if 0 <= choice_idx < len(search_results):
                        return search_results[choice_idx]['symbol']
                    else:
                        print("Invalid selection number. Please try again.")
                except ValueError:
                    print("Please enter a valid number or 'back'")
        else:
            print("No results found or error in search. Please try again.")

def display_realtime_data(symbol, api_key):
    """Display real-time stock data until the user stops it"""
    print(f"\nStarting real-time updates for {symbol} (press Ctrl+C to stop)...")
    try:
        while True:
            data = get_stock_data(symbol, api_key)

            if data:
                print("\nStock Information:")
                print(f"Symbol: {symbol}")
                print(f"Datetime: {data.get('datetime', 'N/A')}")
                print(f"Open: {data.get('open', 'N/A')}")
                print(f"High: {data.get('high', 'N/A')}")
                print(f"Low: {data.get('low', 'N/A')}")
                print(f"Close: {data.get('close', 'N/A')}")
                print(f"Volume: {data.get('volume', 'N/A')}")
            else:
                print("Unable to fetch stock data or invalid symbol")

            sleep(60)  # Wait for 60 seconds before fetching again

    except KeyboardInterrupt:
        print("\nStopped real-time updates.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

def main():
    # Replace with your Twelve Data API key
    API_KEY = 'e163e4023044401db55aa68919412cd5'

    regions = {
        '1': {'name': 'United States', 'exchanges': ['NYSE', 'NASDAQ'], 'example_symbols': ['AAPL', 'MSFT', 'GOOGL']},
        '2': {'name': 'India', 'exchanges': ['BSE', 'NSE'], 'example_symbols': ['RELIANCE', 'TCS', 'INFY']},
        '3': {'name': 'United Kingdom', 'exchanges': ['LSE'], 'example_symbols': ['BP.', 'HSBA', 'VOD']}
    }

    try:
        while True:
            print("\nAvailable Regions:")
            for key, region in regions.items():
                print(f"{key}. {region['name']}")

            region_choice = get_valid_input(
                "\nSelect a region (enter number): ",
                regions.keys()
            )

            selected_region = regions[region_choice]
            print(f"\nSelected Region: {selected_region['name']}")
            print("Major Exchanges:", ", ".join(selected_region['exchanges']))
            print("\nExample Symbols:", ", ".join(selected_region['example_symbols']))

            while True:
                print("\nYou can:")
                print("1. Enter a known stock symbol")
                print("2. Search for a company")
                print("3. Go back to region selection")

                choice = get_valid_input("\nEnter your choice (1, 2, or 3): ", ['1', '2', '3'])

                if choice == '3':
                    break

                symbol = None
                if choice == '1':
                    symbol = input("Enter stock symbol (or 'back' to return): ").upper().strip()
                    if symbol.lower() == 'back':
                        continue
                else:  # choice == '2'
                    symbol = handle_symbol_search(API_KEY)
                    if not symbol:
                        continue

                # Display real-time stock updates
                display_realtime_data(symbol, API_KEY)

    except KeyboardInterrupt:
        print("\nProgram terminated by user.")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")
        print("Please try again.")

if __name__ == "__main__":
    main()
```
